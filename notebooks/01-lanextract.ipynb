```python
import langextract as lx
from dotenv import load_dotenv
import os
load_dotenv() 

pdf_file = "./data/parasanth narasimhan.pdf"

examples = [
    lx.data.ExampleData(
        text="John Doe - Software Engineer with 5 years of experience in Python, JavaScript. Worked at TechCorp. Education: BSc Computer Science, XYZ University.",
        extractions=[
            lx.data.Extraction(
                extraction_class="name",
                extraction_text="John Doe"
            ),
            lx.data.Extraction(
                extraction_class="job_title",
                extraction_text="Software Engineer"
            ),
            lx.data.Extraction(
                extraction_class="experience_years",
                extraction_text="5 years"
            ),
            lx.data.Extraction(
                extraction_class="skills",
                extraction_text="Python, JavaScript"
            ),
            lx.data.Extraction(
                extraction_class="company",
                extraction_text="TechCorp"
            ),
            lx.data.Extraction(
                extraction_class="education",
                extraction_text="BSc Computer Science, XYZ University"
            )
        ]
    )
]
# Run extraction (uses Gemini Flash by default)
result = lx.extract(
    text_or_documents=pdf_file,
    prompt_description="Extract name, job title, skills, experience, education, and companies.",
    examples=examples,
    model_id="gemini-3-flash-preview",
    api_key=os.environ.get("LANGEXTRACT_API_KEY")
)

print(result.extractions)

```
